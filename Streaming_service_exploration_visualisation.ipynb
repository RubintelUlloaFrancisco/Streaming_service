```python
# 2. Plot for movie duration 
# Data preperation for converting the duration for movies from string to numerical values
```

```python
# In the below piece of code my target is to convert the the duration for movies that is given
# in string format into int format.

movie_duration = streaming_service_df_copy[streaming_service_df_copy["type"] == "Movie"]
# In above code I have filter out the data that contains only movies.

df_movie_duration = movie_duration["duration"].apply(lambda x:str(x).split(" ")).to_list()
# In the above line of code I have created a list that has its elements as list 
# which contains numerical value and min seperated by " ,"  per title.

df_movie_duration = pd.DataFrame(df_movie_duration , index = movie_duration["title"])
# In the above line of code I have created a dataframe with index as title from the 
# df_movie_duration. 

df_movie_duration.reset_index(inplace =True)
# Here I have reset the index.

df_movie_duration = df_movie_duration[["title",0]]
# In above line of code I have filtered out the required columns.

df_movie_duration.columns = ["title" ,"duration" ]
# Here i have changed the name of the column as required.

df_movie_duration["duration"]  = df_movie_duration["duration"].astype(float)
# In above line of code I have converted the duration which is in string format to float as 
# nan values cannot be directly converted to int.

df_movie_duration["duration"].fillna(value = 0 , inplace =True)
# now i have replaced duration having NaN with 0.

df_movie_duration["duration"]  = df_movie_duration["duration"].astype(int)
# finally I have converted the duration from float to int datatype 

df_movie_duration_copy = df_movie_duration.copy()
# here I have created a copy

df_movie_duration_copy.drop(columns = ["duration"] , inplace = True)
# now I have dropped the duration column so that i could merge the duration converted in integer form.

df_movie_duration_copy.reset_index(drop= True,inplace=True)
# reset the index

df_movie_duration_copy = df_movie_duration.merge(df_movie_duration_copy , on = "title" , how = "inner")
# now i have merged with durtion that i have converted into int format

# creating graph for above data

# 1. creating histplot ...............

print("HISTPLOT")

plt.figure(figsize = (25,15))
# seeting figure size

sns.histplot(data = df_movie_duration_copy,
            y = "duration",
            bins = 10)

plt.xlabel("Frequency" , fontsize = 15 , labelpad = 15)
plt.ylabel("Duration" , fontsize = 15 , labelpad = 15)
# setting up labels

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
# setting up x and y axis values

plt.show()

# 2. creating boxplot for the above ................

print("BOXPLOT")
plt.figure(figsize = (20,10))

sns.boxplot(data = df_movie_duration_copy,
            x = "duration",
            )

plt.xlabel("Duration" , fontsize = 15 , labelpad = 15)
# setting up labels

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
# setting up x and y axis values

plt.show()
```

NaN value cannot be directly converted to int value reason:

NaN values cannot be directly converted to an integer datatype in pandas dataframe because NaN stands for "Not a Number" and it is a special floating-point value used to represent missing or undefined values. 
In Python, NaN is a float type and cannot be converted directly to an integer datatype. If you try to convert a NaN value to an integer using the int() function, it will raise a ValueError.
To handle missing values in pandas dataframe, you can either replace them with a default value using the fillna() method or drop them using the dropna() method before converting the datatype to integer.

```python
def most_productive_director(x):
    if x["release_year"].min() == x["release_year"].max():
        x["active_year"] = 1 
    else:
        x["active_year"] = x["release_year"].max() - x["release_year"].min()
        
    x["title_count"] = x["title"].nunique()
    x["movies_per_year"] = (x["title_count"] / x["active_year"]).astype(int)
    
    return x

most_pro_director = streaming_service_new_df.groupby("director").apply(most_productive_director)
```

```python
most_pro_director = most_pro_director.loc[most_pro_director["director"]!= "nan"]
most_pro_director = most_pro_director.groupby("director")["movies_per_year"].unique().sort_values(ascending = False)
```
